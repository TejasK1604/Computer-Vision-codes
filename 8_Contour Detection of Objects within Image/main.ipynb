# **Aim:** To detect and visualize the contours of objects within an image

# **Theory**

## **Contours:**
Contours are said to be a curve that simply joins all the continuous points, along the boundary, having the same color or intensity. It can be represented using chain code, Fourier descriptors, shape context etc.

Contours can be defined as the curves or outlines that represent the boundaries of objects or shapes within an image. These curves join all the continuous points (along the boundary) having the same color or intensity, highlighting the structural properties of objects and providing a way to extract and represent shape information from images.

Contours are useful when working with grayscale or binary images where objects are clearly distinguished from the background based on variations in brightness or color. Even with simple images we transform it into grayscale or binary first to detect contours in a simple way.


image.png


## **Contour Detection:**
The process of finding contours in an image involves the following steps:
1. **Edge Detection:** Edges are identified in the image using edge detection algorithms like Canny edge detection.
2. **Contour Extraction:** Contours are extracted from the detected edges using algorithms like the Suzuki algorithm.
3. **Contour Hierarchy:** The hierarchical relationship between contours is established, indicating which contours are enclosed within others.


image.png


## **Applications of Contour Detection:**
- **Object Recognition:** Contours can be used to identify and classify objects in an image.
- **Shape Analysis:** The shape of objects can be analyzed using contour features like area, perimeter, and moments.
- **Object Tracking:** Contours can be used to track the movement of objects in a video sequence.
- **Image Segmentation:** Contours can be used to segment an image into different regions based on object boundaries.

## **Contour Visualization:**
- Contours can be visualized by drawing lines or curves along their boundaries on the original image.
- Different colors or line styles can be used to represent different contours.
- Additional information like contour area, perimeter, or centroid can be displayed along with the contours.

## **Libraries for Contour Detection and Visualization:**
- **OpenCV:** A popular computer vision library that provides functions for contour detection and visualization.
- **Scikit-image:** Another image processing library that offers contour detection capabilities.
- **Matplotlib:** A plotting library that can be used to visualize contours.

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
```

```python
img = cv2.imread("/content/drive/MyDrive/CVR Lab/cont_new.jpeg")
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
binary_img = cv2.threshold(gray_img, 136, 255, cv2.THRESH_BINARY)[1]
plt.imshow(binary_img, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(binary_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours
```

```python
contours_img = cv2.drawContours(img, contours, -1, (0, 255, 0), 4)
plt.imshow(contours_img)
```
